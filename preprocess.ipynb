```python
income_translators = {
    'Δουλεια':'Job',
    'Επίδομα στέγασης': 'Allowance',
    'Αλλα': 'Others',
    'Νταντι': 'Dad',
    'Λενα': 'Lena',
    'Ιδιαίτερο': 'Private',
    'Mam': 'Mom',
    'Δώρο': 'Gift'
}
```

```python
expenses_translators = {
    'Καφετέρια': 'Fun',
    'Είδη παντοπωλείου': 'Supermarket',
    'Προπόνηση': 'Gym',
    'Βετζινα': 'Fuel',
    'Υγεία': 'Health',
    'Κουρεμα': 'Haircut',
    'Πάρτυ μ': 'Me',
    'Ελεύθερος χρόνος': 'Hobies',
    'Ρευμα': 'Electricity',
    'Κοινόχρηστα': 'Building fees',
    'Αλλα': 'Other',
    'Συνδρομες': 'Subscriptions',
    'Vodafone': 'Vodafone',
    'Νοικι': 'Rent',
    'Δώρα': 'Gifts',
    'Μπρικης': 'Mprikis',
    'Εκπαίδευση': 'Education',
    'Διακοπες': 'Vacation',
    'Οικογένεια': 'Family',
    'Ευδαπ': 'Water',
    'Παρκινγκ': 'Parking',
    'Καρτα': 'Vodafone',
    'Σέρβις/λαδια/ασφαλειες': 'MOTOmaintenance/oil/insurance',
    'Λογιστης': 'Accountant',
    'Εφκα': 'Efka',
    'Διακοπές παρεα': 'Vacation'
}
```

```python
common_df = common_df.drop(columns=['Προεπιλεγμένο νόμισμα','Ποσό σε νόμισμα λογαριασμού','Νόμισμα Λογαριασμού','Ποσό συναλλαγής σε νόμισμα συναλλαγής','Νόμισμα συναλλαγής','Ετικέτες','Σχόλιο','Λογαριασμός'])
common_df = common_df.rename(columns={'Ημερομηνία και ώρα': 'Date'})
common_df = common_df.rename(columns={'Κατηγορία': 'Category'})
common_df = common_df.rename(columns={'Ποσό σε προεπιλεγμένο νόμισμα': 'Amount'})
common_df  
```

```python
common_df['Total'] = ''
common_df['Type'] = 0
common_df
```

```python
# Function to translate and update type
def translate_category(row):
    category = row['Category']
    if category in expenses_translators:
        return expenses_translators[category], 0
    elif category in income_translators:
        return income_translators[category], 1
    else:
        return category, row['Type']  # Keep original if not found

# Apply the translation and type update
common_df['Category'], common_df['Type'] = zip(*common_df.apply(translate_category, axis=1))
```

```python
common_df = common_df.sort_values(by=['Date'], ascending=False).reset_index(drop=True)
common_df
```

```python
common_df.to_csv('common_df.csv', index=False)
```

```python
final_df = pd.read_csv('balance_24_25 - in_out.csv')
final_df = final_df.sort_values(by=['Date'], ascending=False).reset_index(drop=True)
final_df
```

```python
# Calculate running total with proper handling of income/expenses
def calculate_running_total(df):
    running_total = 0
    totals = []
    
    for _, row in df.iterrows():
        amount = row['Amount']
        if row['Type'] == 1:  # Income
            running_total += amount
        else:  # Expense
            running_total -= amount
        totals.append(round(running_total, 1))  # Round to 1 decimal place
    
    return totals

# Calculate and update the totals
final_df['Total'] = calculate_running_totals(final_df)
```
